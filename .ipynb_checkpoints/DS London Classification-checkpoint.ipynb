```python
# importing library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
trainlabel = pd.read_csv('trainLabels.csv')
```

```python
print(train.shape)
train.head(2)
```

```python
print(trainlabel.shape)
trainlabel.head(2)
```

```python
print(test.shape)
test.head(2)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train, trainlabel, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
print(f"Logistic Regression Accuracy Score : {accuracy_score(y_pred,y_test)}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier, BaggingClassifier
from catboost import CatBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

def classification_model(x,y):
    models={
        'Logistic Regression': LogisticRegression(),
        'Decision Tree': DecisionTreeClassifier(),
        'Random Forest': RandomForestClassifier(),
        'AdaBoost' : AdaBoostClassifier(),
        'Gradient Boost' : GradientBoostingClassifier(),
        'XGBoost' : XGBClassifier(),
        'Voting' : VotingClassifier(estimators=[('DT', DecisionTreeClassifier()),('RF', RandomForestClassifier())]),
        'Bagging' : BaggingClassifier(),
        # 'CatBoost' : CatBoostClassifier(),
        'Gaussian' : GaussianNB(),
        'KNN' : KNeighborsClassifier(),
        "SVM" : SVC()
    }
    X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
    
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred=model.predict(X_test)

        acc = accuracy_score(y_pred, y_test)

        print(f"Model : {name}")
        print(f"Accuracy Score : {acc}")
        print("++"*50)
```

```python
classification_model(train, trainlabel)
```

```python
voting_model = VotingClassifier(estimators=[
    ('DecisionTree', DecisionTreeClassifier()),
    ('RandomForest', RandomForestClassifier()),
    ('KNN', KNeighborsClassifier())
], voting='hard')

voting_model.fit(X_train, y_train)
```

```python
y_pred_voting = voting_model.predict(X_test)

acc_score_voting = accuracy_score(y_pred_voting, y_test)
print(f"Accuracy Score using Voting Classifier : {acc_score_voting}")
```

```python
# Scaling the dataset

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
pca = PCA()

X_train_sc = scaler.fit_transform(X_train)
y_train_sc = scaler.fit(y_train)

X_train_pca = pca.fit_transform(X_train_sc)
y_train_pca = pca.fit(X_train_sc)
```

```python
classification_model(X_train_pca, y_train)

```
