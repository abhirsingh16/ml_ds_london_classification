```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train, trainlabel, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train.values.ravel())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier, BaggingClassifier
from catboost import CatBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

def classification_model(x,y):
    models={
        'Logistic Regression': LogisticRegression(),
        'Decision Tree': DecisionTreeClassifier(),
        'Random Forest': RandomForestClassifier(),
        'AdaBoost' : AdaBoostClassifier(),
        'Gradient Boost' : GradientBoostingClassifier(),
        'XGBoost' : XGBClassifier(),
        'Voting' : VotingClassifier(estimators=[('DT', DecisionTreeClassifier()),('RF', RandomForestClassifier())]),
        'Bagging' : BaggingClassifier(),
        # 'CatBoost' : CatBoostClassifier(),
        'Gaussian' : GaussianNB(),
        'KNN' : KNeighborsClassifier(),
        "SVM" : SVC()
    }
    X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
    
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred=model.predict(X_test)

        acc = accuracy_score(y_pred, y_test)

        print(f"Model : {name}")
        print(f"Accuracy Score : {acc}")
        print("++"*50)
```

```python
classification_model(X_train, y_train.values.ravel())
```

```python
voting_model = VotingClassifier(estimators=[
    ('DecisionTree', DecisionTreeClassifier()),
    ('RandomForest', RandomForestClassifier()),
    ('KNN', KNeighborsClassifier())
], voting='hard')

voting_model.fit(X_train, y_train.values.ravel())
```

```python
y_pred_voting = voting_model.predict(X_test)

acc_score_voting = accuracy_score(y_pred_voting, y_test)
print(f"Accuracy Score using Voting Classifier : {acc_score_voting}")
```

```python
# print(f"Model: {cross_val_score(KNN, train_norm, trainlabel.values.ravel(), cv=10).mean()}")

def cross_val_classification_model(x,y):
    models={
        'Logistic Regression': LogisticRegression(),
        'Decision Tree': DecisionTreeClassifier(),
        'Random Forest': RandomForestClassifier(),
        'AdaBoost' : AdaBoostClassifier(),
        'Gradient Boost' : GradientBoostingClassifier(),
        'XGBoost' : XGBClassifier(),
        'Bagging' : BaggingClassifier(),
        'Gaussian' : GaussianNB(),
        'KNN' : KNeighborsClassifier(),
        "SVM" : SVC(),
        "Voting" : VotingClassifier(estimators=[
        ('DecisionTree', DecisionTreeClassifier()),
        ('RandomForest', RandomForestClassifier()),
        ('KNN', KNeighborsClassifier())
        ])
    }

    for model_name, model in models.items():
        cv_score = cross_val_score(model, x, y, cv=10)
        print(f"Model Name : {model_name}")
        print(f"Cross Validataion Score : {cv_score.mean():.4f}")
        print(f"=="*40)
```

```python
cross_val_classification_model(train, trainlabel.values.ravel())
```

```python
from sklearn.decomposition import PCA

pca = PCA()

pca.fit(train)

# Get the explained variance ratio for each component
explained_variance = pca.explained_variance_ratio_

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(explained_variance)


# Display the cumulative explained variance to inspect
print("Cumulative Explained Variance")
for i, variance in enumerate(cumulative_variance):
    print(f"Component {i+1} : {variance:.4f}")


# Find the number of components that explain at least 80% and at most 95% of the variance
n_components_80 = np.argmax(cumulative_variance >= 0.80)+1   # Adding 1 because index starts from 0
n_components_90 = np.argmax(cumulative_variance >= 0.90)+1

print(f"Number of Components that explain 80% of the variance : {n_components_80}")
print(f"Number of Components that expalin 90% of the variance : {n_components_90}")
```

```python
pca = PCA(n_components=n_components_80)

train_pca = pca.fit_transform(train)
```

```python
train_pca.shape
```

```python
cross_val_classification_model(train_pca, trainlabel.values.ravel() )
```

#### Applying Gaussian Mixture and Grid Search to improve accuracy

```python
from sklearn.model_selection import GridSearchCV
from sklearn.mixture import GaussianMixture
```

```python
x_all = np.r_[train, test]
x_all.shape
```

##### Applying Gaussian Mixture

```python
lowest_bic = np.infty
bic = []
n_components_range = range(1, 7)
cv_types = ['spherical', 'tied', 'diag', 'full']
for cv_type in cv_types:
    for n_components in n_components_range:
        gmm = GaussianMixture(n_components=n_components,covariance_type=cv_type)
        gmm.fit(x_all)
        bic.append(gmm.aic(x_all))
        if bic[-1] < lowest_bic:
            lowest_bic = bic[-1]
            best_gmm = gmm
            
best_gmm.fit(x_all)
gmm_train = best_gmm.predict_proba(train)
gmm_test = best_gmm.predict_proba(test)

#Random Forest Classifier
rfc = RandomForestClassifier(random_state=99)

#USING GRID SEARCH
n_estimators = [10, 50, 100, 200,400]
max_depth = [3, 10, 20, 40]
param_grid = dict(n_estimators=n_estimators,max_depth=max_depth)

grid_search_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv = 10,scoring='accuracy',n_jobs=-1).fit(gmm_train, trainlabel.values.ravel())
rfc_best = grid_search_rfc.best_estimator_
print('Random Forest Best Score',grid_search_rfc.best_score_)
print('Random Forest Best Parmas',grid_search_rfc.best_params_)
print('Random Forest Accuracy',cross_val_score(rfc_best,gmm_train, trainlabel.values.ravel(), cv=10).mean())

#KNN 
knn = KNeighborsClassifier()

#USING GRID SEARCH
n_neighbors=[3,5,6,7,8,9,10]
param_grid = dict(n_neighbors=n_neighbors)

grid_search_knn = GridSearchCV(estimator=knn, param_grid=param_grid, cv = 10, n_jobs=-1,scoring='accuracy').fit(gmm_train,trainlabel.values.ravel())
knn_best = grid_search_knn.best_estimator_
print('KNN Best Score', grid_search_knn.best_score_)
print('KNN Best Params',grid_search_knn.best_params_)
print('KNN Accuracy',cross_val_score(knn_best,gmm_train, trainlabel.values.ravel(), cv=10).mean())

#SVM
svc = SVC()

#USING GRID SEARCH
parameters = [{'kernel':['linear'],'C':[1,10,100]},
              {'kernel':['rbf'],'C':[1,10,100],'gamma':[0.05,0.0001,0.01,0.001]}]
grid_search_svm = GridSearchCV(estimator=svc, param_grid=parameters, cv = 10, n_jobs=-1,scoring='accuracy').fit(gmm_train, trainlabel.values.ravel())
svm_best = grid_search_svm.best_estimator_
print('SVM Best Score',grid_search_svm.best_score_)
print('SVM Best Params',grid_search_svm.best_params_)
print('SVM Accuracy',cross_val_score(svm_best,gmm_train, trainlabel.values.ravel(), cv=10).mean())
```
